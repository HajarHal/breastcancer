```python
import networkx as nx
import numpy as np

# Create a directed graph
G = nx.DiGraph()

# Add edges (relationships) to the graph with initial weights
for index, row in df.iterrows():
    G.add_edge(row['SUBJECT_NAME'], row['OBJECT_NAME'], weight=1)  # Initial weight can be set to 1 or some heuristic

# Apply a random walk with restart
def random_walk_with_restart(graph, start_node, alpha=0.85, max_iter=100):
    # Initialize probability distribution
    n = len(graph)
    prob_dist = np.zeros(n)
    prob_dist[start_node] = 1.0

    # Adjacency matrix
    A = nx.to_numpy_array(graph)
    degrees = np.sum(A, axis=1)
    
    # Handle zero-degree nodes to avoid division by zero
    with np.errstate(divide='ignore', invalid='ignore'):
        P = np.divide(A, degrees[:, None], where=degrees[:, None] != 0)
        P[degrees == 0] = 0

    for _ in range(max_iter):
        prob_dist = (1 - alpha) * np.dot(P, prob_dist) + alpha * np.zeros(n)
        prob_dist[start_node] = 1.0  # Restart condition

    return prob_dist

# Map node names to indices
node_index = {node: i for i, node in enumerate(G.nodes())}
index_node = {i: node for node, i in node_index.items()}

# Apply the random walk to each node in the graph
factuality_scores = {}
for node in G.nodes():
    start_node_index = node_index[node]
    score_distribution = random_walk_with_restart(G, start_node_index)
    factuality_scores[node] = np.mean(score_distribution)  # You can aggregate scores differently

# Assign scores to the relationships in the original dataset
df['factuality_score'] = df.apply(lambda row: (factuality_scores[row['SUBJECT_NAME']] +
                                               factuality_scores[row['OBJECT_NAME']]) / 2, axis=1)

# Save the updated DataFrame with factuality scores
output_file_path = 'tt.csv' 
df.to_csv(output_file_path, index=False)

print(f"Les scores de factualité ont été ajoutés et sauvegardés dans '{output_file_path}'.")
```

```python
import random
import pandas as pd

import networkx as nx

G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['SUBJECT_NAME'], row['OBJECT_NAME'], relation=row['PREDICATE'])

def random_walk(graph, start_node, num_steps):
    current_node = start_node
    visited_nodes = {current_node: 1}
    
    for _ in range(num_steps):
        neighbors = list(graph.neighbors(current_node))
        if not neighbors:
            break
        next_node = random.choice(neighbors)
        if next_node in visited_nodes:
            visited_nodes[next_node] += 1
        else:
            visited_nodes[next_node] = 1
        current_node = next_node
    
    return visited_nodes

def calculate_factuality_scores(df, num_steps=1000):
    scores = {}
    
    subject_nodes = df['SUBJECT_NAME'].unique()
    
    for subject in subject_nodes:
        visited = random_walk(G, subject, num_steps)
        
        for _, row in df.iterrows():
            subject = row['SUBJECT_NAME']
            obj = row['OBJECT_NAME']
            total_visits = sum(visited.values())
            score = visited.get(obj, 0) / total_visits if total_visits > 0 else 0
            scores[(subject, obj)] = score
    
    return scores

factuality_scores = calculate_factuality_scores(df)


df['Factuality_Score'] = df.apply(lambda row: factuality_scores.get((row['SUBJECT_NAME'], row['OBJECT_NAME']), 0), axis=1)

output_file_path = 'score_data.csv' 
df.to_csv(output_file_path, index=False)

print(f"Les scores de factualité ont été ajoutés et sauvegardés dans '{output_file_path}'.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load the dataset
df = pd.read_csv("cause_dataset.csv", encoding='latin1')

# Combine SUBJECT_NAME and OBJECT_NAME into a single text column
df['text'] = df['SUBJECT_NAME'] + " " + df['OBJECT_NAME']

X = df['text']
y = df['Category']

# Separate known and unknown labels
known_labels_df = df[df['Category'] != 'Other']
unknown_labels_df = df[df['Category'] == 'Other']

# Split the known labels into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    known_labels_df['text'], known_labels_df['Category'], test_size=0.3, random_state=42
)

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Initialize and train the Decision Tree classifier
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train_tfidf, y_train)

# Make predictions
y_pred = model.predict(X_test_tfidf)

# Evaluate accuracy, precision, recall, and F1-score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Print the metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Predict categories for the unknown labels
X_unknown_tfidf = vectorizer.transform(unknown_labels_df['text'])
unknown_labels_df['Category'] = model.predict(X_unknown_tfidf)

# Combine the known and classified unknown labels
df_combined = pd.concat([known_labels_df, unknown_labels_df], axis=0)
df_combined.to_csv("classified_data.csv", index=False)
```

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier CSV dans un DataFrame
file_path = 'classified_data.csv' 
df = pd.read_csv(file_path)

# Filtrer les données pour ne conserver que les facteurs génétiques et le SUBJECT_NAME spécifique
df_genetic_her2 = df[(df['Category'] == 'Hormone') & (df['Cancer_Type'] == 'Male Breast Cancer')]

# Créer le graphe pour les facteurs génétiques de HER2-negative breast cancer
G_genetic_her2 = nx.Graph()
for _, row in df_genetic_her2.iterrows():
    G_genetic_her2.add_edge(row['SUBJECT_NAME'], row['Cancer_Type'], relation=row['Factuality_Score'])

# Visualiser le graphe
plt.figure(figsize=(11, 5))
pos = nx.spring_layout(G_genetic_her2, seed=42) 

# Dessiner les nœuds et les arêtes
nx.draw(G_genetic_her2, pos, with_labels=True, node_color='lightblue', node_size=1000, edge_color='gray', font_size=8, font_weight='bold', alpha=0.9)

# Dessiner les étiquettes des arêtes
edge_labels = nx.get_edge_attributes(G_genetic_her2, 'relation')
nx.draw_networkx_edge_labels(G_genetic_her2, pos, edge_labels=edge_labels)


# Enregistrer l'image du graphe
plt.savefig('static/genetic_cancer_graph.png')

# Afficher l'image
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les classified_data
df = pd.read_csv('classified_data.csv')

# Calculer la fréquence des causes par type de cancer
frequency_df = df.groupby(['Cancer_Type', 'Category']).size().reset_index(name='Frequency')

# Exemple de filtrage pour le cancer du sein masculin
male_cancer_causes = frequency_df[frequency_df['Cancer_Type'] == 'Male Breast Cancer']

# Exemple de condition pour colorer les barres
threshold = male_cancer_causes['Frequency'].mean()  # Utilisation de la moyenne comme seuil
colors = male_cancer_causes['Frequency'].apply(lambda x: 'red' if x > threshold else '#3fbbc0')

# Création du graphique avec des barres de couleur par défaut
plt.figure(figsize=(8, 6))
barplot = sns.barplot(x='Category', y='Frequency', data=male_cancer_causes, palette=['#3fbbc0']*len(male_cancer_causes))

# Application des couleurs conditionnelles aux barres
for i, bar in enumerate(barplot.patches):
    bar.set_color(colors.iloc[i])

# Rotation des étiquettes
plt.xticks(rotation=90)

# Supprimer l'étiquette du label 'Category'
barplot.set(xlabel='')
# Supprimer l'étiquette du label 'Category'
barplot.set(ylabel='')

# Ajustement du layout pour éviter la coupure des étiquettes
plt.tight_layout()

# Sauvegarde du graphique
plt.savefig('assets/stat/causes_male/male_cancer_causes_frequency.png')

# Affichage du graphique
plt.show()
```

```python
from pyvis.network import Network
import pandas as pd
import os

file_path = 'classified_data.csv'
df = pd.read_csv(file_path)

output_dir = 'C:/Users/PcPack/cancer/assets/graph/causes_male'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
categories = df['Category'].unique()
for category in categories:
    df_category = df[(df['Category'] == category) & (df['Cancer_Type'] == 'Male Breast Cancer')]
    net = Network(notebook=True)
    for _, row in df_category.iterrows():
        formatted_score = f"{row['Factuality_Score']:.7f}"
        net.add_node(row['SUBJECT_NAME'], title=row['SUBJECT_NAME'], size=15, color='lightblue')
        net.add_node(row['Cancer_Type'], title=row['Cancer_Type'], size=15, color='lightblue')
        net.add_edge(
            row['SUBJECT_NAME'], 
            row['Cancer_Type'], 
            title=f"Factuality Score: {formatted_score}", 
            label=formatted_score,  
            color='gray'
        )
    file_path = os.path.join(output_dir, f'{category}.html')
    net.show(file_path)
```
